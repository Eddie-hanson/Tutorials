```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X= np.arange(0,100)
Y=X*2
Z=X**2
```

```python
X
```

```python
from matplotlib.axes import Axes
fig=plt.figure()
Ax=fig.add_axes([0,0,1,1])
Ax.plot(X,Y)
Ax.set_title('My Plot')
Ax.set_xlabel('X')
Ax.set_ylabel('Y')
```

```python
fig2=plt.figure()
Ax1=fig2.add_axes([0,0,1,1])
Ax2=fig2.add_axes([0.2,0.5,0.2,0.2])

```

```python
fig2=plt.figure()

Ax1=fig2.add_axes([0,0,1,1])
Ax1.plot(X,Y)
Ax1.set_xlabel('X')
Ax1.set_ylabel('Y')

Ax2=fig2.add_axes([0.2,0.5,0.2,0.2])  
Ax2.plot(X,Y, color='red')
Ax2.set_xlabel('X')
Ax2.set_ylabel('Y')
```

```python
fig3=plt.figure()

Ax3=fig3.add_axes([0,0,1,1])
Ax4=fig3.add_axes([0.2,0.5,0.4,0.4])
```

```python
fig3=plt.figure()

Ax3=fig3.add_axes([0,0,1,1])
Ax3.plot(X,Y,Z)
Ax3.set_title('Original')
Ax3.set_xlabel('X')
Ax3.set_ylabel('Z')

Ax4=fig3.add_axes([0.2,0.5,0.4,0.4])
Ax4.set_title('Zoom')
Ax4.set_xlabel('X')
Ax4.set_ylabel('Y')
Ax4.plot(X,Y,Z)
Ax4.set_xlim([20,22])
Ax4.set_ylim([300,500])
```

```python
fig, Ax=plt.subplots(nrows=1,ncols=2)
for A in Ax:
    A.plot()
```

```python
fig ,Ax=plt.subplots(nrows=1, ncols=2)
Ax[0].plot(X,Y, color='blue', lw='2', ls='--')
Ax[1].plot(X,Z, color='red', lw=3)

```

```python
fig ,Ax=plt.subplots(nrows=1, ncols=2 ,figsize=(12,2))
Ax[0].plot(X,Y, color='blue', lw='2')
Ax[1].plot(X,Z, color='red', lw=3,  ls='--')
```
