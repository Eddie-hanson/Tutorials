```python
#import matplotlib as plt
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
X=np.linspace(0,5,11)
Y = X**2
```

```python
X
```

```python
Y
```

# Funtional method of creating a plot in matplotlib

```python
plt.plot(X,Y)
plt.xlabel('X-Values')
plt.ylabel('Y-Values')
plt.title('Visualization')
```

```python
plt.subplot(1,2,1)
plt.plot(X,Y, 'r')

plt.subplot(1,2,2)
plt.plot(X,Y, 'g')
```

## Object Oriented Method

```python
fig=plt.figure()
Ax=fig.add_axes([0.1,0.1,0.8,0.8])
Ax.plot(X,Y)
Ax.set_xlabel('X axis')
Ax.set_ylabel('Y-axis')
Ax.set_title('OOP')
```

```python
fig=plt.figure()
Axes1=fig.add_axes([0.1,0.1,0.8,0.8])
Axes2=fig.add_axes([0.2,0.5,0.4,0.3])

Axes1.plot(X,Y)
Axes1.set_xlabel('X1 axis')
Axes1.set_ylabel('Y1 axis')
Axes1.set_title('My X1 plot')

Axes2.plot(Y,X)
Axes2.set_xlabel('X2 axis')
Axes2.set_ylabel('Y2 axis')
Axes2.set_title('My X2 plot')
```

```python
fig,Ax=plt.subplots(nrows=1,ncols=2)
# Ax.plot(X,Y)
for i in Ax:
    i.plot(X,Y)
```

```python
fig,Ax=plt.subplots(nrows=1,ncols=2)
Ax[0].plot(X,Y)
Ax[0].set_xlabel('AX1')
Ax[0].set_ylabel('AY1')
Ax[0].set_title('First Plot')

```

```python
fig,Ax=plt.subplots(nrows=1,ncols=2)
Ax[1].plot(Y,X)
Ax[1].set_xlabel('AX2')
Ax[1].set_ylabel('AY2')
Ax[1].set_title('Second Plot')
```

```python
fig=plt.figure(figsize=(8,4))
Axes=fig.add_axes([0,0,0.8,0.7])
Axes.plot(Y,X)
```

```python
fig,Axes=plt.subplots(figsize=(8,4), nrows=1, ncols=2)
Axes[0].plot(X,Y)
Axes[0].set_xlabel('X-axes1')
Axes[0].set_ylabel('Y-axis1')
Axes[0].set_title('First Plot')

Axes[1].plot(X,Y**2, label='Y squared')
Axes[1].plot(X,X**2, label='X squared')
Axes[1].set_xlabel('X-axes2')
Axes[1].set_ylabel('Y-axis2')
Axes[1].set_title('Second Plot')
Axes[1].legend(loc=10)
plt.tight_layout
```
