# SUBSETTING DATA
Having a data with too many columns and you want to work with or subset to a few columns you want to work with

### SUBSETTING COLUMNS

```python
"""Lists all columns available in a Dataframe
    """

df.columns
```

```python
"""df=df[['Name1','Name2','Name3','Name4','Name5']] is used to subset a list of columns you want to keep 
and work with in your dataframe.
    """
df[['title', 'description', 'publishTime', 'duration_seconds',
       'viewCount', 'likeCount', 'commentCount']]
```

```python
"""Setting subset as your new dataframe
    """
df=df[['title', 'description', 'publishTime', 'duration_seconds',
       'viewCount', 'likeCount', 'commentCount']]
```

### SUBSETTING ROWS

```python
"""viewing rows where viewcount is > a certtain number
    """
df['viewCount'] > 1_000_000
```

```python
"""subsetting rows where viewcount is > a certtain number
Subsetting usng loc
    """
df_subset1=df.loc[df['viewCount'] > 1_000_000]
df_subset1
```

```python
"""df.query() to filter out row on our dataframe"""
df_subset2=df.query ('likeCount > 1000000')
df_subset2
```

```python
df
```

```python
"""Run .isna on a columns ie df['NameOfColumn'].isna() to 
view true is a value in a column is a no value ie NaN or has no value
"""
df["likeCount"]
```

```python
df["likeCount"].isna()
```

```python
""" ~df['NameOfColumn'].isna() to 
view true in columns with values.
"""
~df["likeCount"].isna()
```

```python
""""Subsetting data to to include columns with values
"""
df.loc[~df["likeCount"].isna()]
```

```python
""""   New dataframe
    """
df=df.loc[~df["likeCount"].isna()]
df
```

# Casting Dtypes of a series in s dataframe
Changing a series into a different type that it already is. 

```python
df.dtypes
```

```python
"""Changing a columns dtype to another dtype use .astype('NameOfDataType (Int,Float, Object, etc)')
or pd.to_numeric(df['NameOfColumn'])
"""
df['likeCount'].astype('int')
df['viewCount'].astype('int')
```
