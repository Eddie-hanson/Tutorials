```python
#importing numpy
import numpy as np
from cProfile import label
# importing matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
from numpy.random import rand


RandNmber=np.random.rand(10)
RandNmber
```

```python
#style of the plot



style.use('ggplot')
#pllot the random number
plt.plot(RandNmber, 'g',label='line one', linewidth =2)
#x-axis 
plt.xlabel('Range')
#y-axis
plt.ylabel('Number')
#title
plt.title('My Plot')
```

# 2D PLOTS

```python
Users=[123,654,67,883,56,126,43,54,245,66]
Time=[3,6,4,8,14,45,3,28,36,52]
len(Users)
```

```python
#style of the plot



style.use('ggplot')
#pllot the random number
plt.plot(Time,Users, 'g',label='line one', linewidth =2, color= 'c', linestyle='--')
#x-axis 
plt.xlabel('Users')
#y-axis
plt.ylabel('Time')
#title
plt.title('My Plot')
```

```python
Monday=[123,654,67,883,56,126,43,54,245,66]
Tuesday=[13,64,367,183,256,526,243,104,205,166]
Wednesday=[23,254,367,83,516,170,434,514,24,166]
Thursday=[223,354,467,283,556,226,143,254,255,366]
Friday=[623,351,401,88,206,109,143,541,45,616]

Time=[3,6,4,8,14,45,3,28,36,52]
```
